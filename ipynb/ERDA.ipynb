# Running R and RStudio on ERDA Jupyter server

## Sign-up - First time only
All data analysis will be run in R (https://www.r-project.org) through the ERDA servers. If you have not used ERDA before you first have to sign-up by following the sign-up procedure on http://www.erda.dk/. Use your KU credentials to sign-up (xyz123 and associated password). Do not use your email as the username, only the credientials. If you experience problems with logging in to ERDA it usually helps to clear the browser history and try again.

## Import scripts and data
First, login to http://www.erda.dk/

Before each analysis, you need to import the pre-made scripts and the data.

1) **First you create a folder where the scripts and data will go.**
* Right-click in the left box and click Create Folder:
![ERDA1](../img/erda1.png)
* Give it a fitting name, click Ok:
![ERDA2](../img/erda2.png)

2) **Import data.**
* Right-click on your new folder, press Share Link -> Import:
![ERDA3](../img/erda3.png)
* Enter the appropriate Share Link ID ([see absalon for link IDs for the different analyses](https://absalon.ku.dk/courses/47356/pages/data-analysis-share-links)), click Ok:
![ERDA4](../img/erda4.png)


## Start RStudio through ERDA
First, login to http://www.erda.dk/

RStudio is a software which simplifies analyses with R.

To run R-Studio, you first have to start a Jupyter server. This server will have both R, R-Studio, and all relevant R packages pre-installed.

1) **Start your own Jupyter server**
* Click Jupyter in the left menu:
![ERDA5](../img/erda5.png)
* Click Start DAG:
![ERDA6](../img/erda6.png)
* Click Start My Server.
* Under Select a notebook image choose Bio Notebook:
![ERDA7](../img/erda7.png)

* Click Spawn.

* Now you have access to your own server. Here you can run R, RStudio, Python, and much more. If the server is inactive for 2 hours it will be terminated, which means that everything that is not saved will be lost, and you will have to start a new server.

2) **Start RStudio.**
* Click RStudio -> Launch RStudio.
![ERDA8](../img/erda8.png)

3) **Find the imported scripts and data.**
* In the bottom right corner click *work*.
![ERDA9](../img/erda9.png)

* Click the name of the folder you imported the data into.
* Click the name of the script you want to open (E.g. Intro.R).
* You are now ready to start analysing with R!
